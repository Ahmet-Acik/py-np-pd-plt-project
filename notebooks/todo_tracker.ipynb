## Define Todo Item Class
Define a class to represent a todo item with attributes like description, status (pending or completed), and creation date.

```python
class TodoItem:
    def __init__(self, description):
        self.description = description
        self.status = 'pending'
        self.creation_date = datetime.datetime.now()

    def __str__(self):
        return f"[{self.status.upper()}] {self.description} (Created: {self.creation_date})"
```
