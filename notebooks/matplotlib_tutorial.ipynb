# Data Visualization with Matplotlib

Matplotlib is the foundation for data visualization in Python, providing extensive capabilities for creating static, animated, and interactive plots.

## Basic Plotting

```python
import matplotlib.pyplot as plt
import numpy as np

# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='sin(x)')
plt.title('Simple Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Multiple Plots and Subplots

```python
# Multiple functions on same plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(12, 8))

# Subplot 1: Individual plots
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r-', label='sin(x)')
plt.plot(x, y2, 'b-', label='cos(x)')
plt.title('Trigonometric Functions')
plt.legend()
plt.grid(True)

# Subplot 2: Combined plot
plt.subplot(2, 2, 2)
plt.plot(x, y1, 'r--', alpha=0.7, label='sin(x)')
plt.plot(x, y2, 'b--', alpha=0.7, label='cos(x)')
plt.plot(x, y3, 'g-', linewidth=2, label='sin(x)*cos(x)')
plt.title('Combined Functions')
plt.legend()
plt.grid(True)

# Subplot 3: Scatter plot
plt.subplot(2, 2, 3)
random_x = np.random.randn(100)
random_y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(random_x, random_y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar()
plt.title('Random Scatter Plot')
plt.grid(True)

# Subplot 4: Histogram
plt.subplot(2, 2, 4)
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Normal Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Advanced Plotting Features

```python
# Statistical plots
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1.5, 1000)

plt.figure(figsize=(15, 10))

# Box plot
plt.subplot(2, 3, 1)
plt.boxplot([data1, data2], tick_labels=['Dataset 1', 'Dataset 2'])
plt.title('Box Plot Comparison')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)

# Violin plot
plt.subplot(2, 3, 2)
plt.violinplot([data1, data2], showmeans=True, showmedians=True)
plt.xticks([1, 2], ['Dataset 1', 'Dataset 2'])
plt.title('Violin Plot')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)

# Error bars
plt.subplot(2, 3, 3)
x = np.arange(5)
y = np.array([1, 3, 2, 4, 3])
errors = np.array([0.2, 0.3, 0.1, 0.4, 0.2])

plt.errorbar(x, y, yerr=errors, fmt='o-', capsize=5, capthick=2, elinewidth=2)
plt.title('Error Bars')
plt.xlabel('Measurement')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)

# Contour plot
plt.subplot(2, 3, 4)
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, levels=20, cmap='RdYlBu')
plt.colorbar()
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')

# 3D surface plot
from mpl_toolkits.mplot3d import Axes3D

ax = plt.subplot(2, 3, 5, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
plt.colorbar(surf, ax=ax, shrink=0.5)
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Pie chart
plt.subplot(2, 3, 6)
sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#ff99cc']
explode = (0.1, 0, 0, 0, 0)  # explode first slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pie Chart Example')

plt.tight_layout()
plt.show()
```

## Customization and Styling

```python
# Style customization
plt.style.use('seaborn-v0_8')  # or 'ggplot', 'bmh', 'fivethirtyeight', etc.

# Custom styling example
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12, 8))

# Plot with custom styling
plt.plot(x, y1, color='#2E86AB', linewidth=3, linestyle='-', marker='o',
         markersize=4, markerfacecolor='white', markeredgecolor='#2E86AB',
         markeredgewidth=2, label='sin(x)', alpha=0.8)

plt.plot(x, y2, color='#F24236', linewidth=2, linestyle='--', marker='s',
         markersize=4, markerfacecolor='#F24236', markeredgecolor='white',
         markeredgewidth=1, label='cos(x)', alpha=0.8)

# Custom title and labels with LaTeX
plt.title(r'$\sin(x)$ and $\cos(x)$ Functions', fontsize=16, fontweight='bold',
          color='#2E86AB', pad=20)
plt.xlabel('x values', fontsize=12, fontweight='bold')
plt.ylabel('Function values', fontsize=12, fontweight='bold')

# Custom grid
plt.grid(True, which='both', linestyle=':', alpha=0.6, color='gray')

# Custom legend
plt.legend(loc='upper right', fontsize=12, framealpha=0.9, fancybox=True,
           shadow=True, borderpad=1)

# Custom spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)

# Custom ticks
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')

# Add text annotation
plt.annotate('Maximum of sin(x)', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 0.8),
             arrowprops=dict(facecolor='black', shrink=0.05, width=2),
             fontsize=12, ha='center')

plt.tight_layout()
plt.show()
```

## Saving Plots

```python
# Save plots in different formats
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-x/10)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', linewidth=2)
plt.title('Damped Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)

# Save in different formats with high quality
plt.savefig('damped_sine_wave.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.savefig('damped_sine_wave.pdf', dpi=300, bbox_inches='tight')
plt.savefig('damped_sine_wave.svg', dpi=300, bbox_inches='tight')

plt.show()

print("Plots saved as PNG, PDF, and SVG formats!")
```

## Matplotlib Summary

**Key Features:**
- Extensive plotting capabilities (line, scatter, bar, histogram, etc.)
- Full customization of colors, styles, markers, and layouts
- Support for mathematical expressions (LaTeX)
- Multiple output formats (PNG, PDF, SVG, EPS)
- Integration with NumPy arrays
- Object-oriented and pyplot interfaces

**Best Practices:**
- Use `plt.figure(figsize=(width, height))` for consistent sizing
- Always add labels, titles, and legends
- Use `plt.tight_layout()` to prevent overlapping
- Save high-resolution plots with `dpi=300`
- Use stylesheets for consistent appearance
- Leverage subplots for multiple related plots

**Common Plot Types:**
- Line plots: `plt.plot()`
- Scatter plots: `plt.scatter()`
- Bar charts: `plt.bar()`
- Histograms: `plt.hist()`
- Box plots: `plt.boxplot()`
- Contour plots: `plt.contour()`
- 3D plots: `from mpl_toolkits.mplot3d import Axes3D`
